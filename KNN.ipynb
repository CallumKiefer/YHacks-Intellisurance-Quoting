```python
import pandas as pd
import datetime
import numpy as np
from sklearn import neighbors as knn
```

```python
data = pd.read_csv("extendedData.csv")
data.columns
data = data.drop('Unnamed: 0', 1)
data = data.drop('address', 1)
data = data.drop('id', 1)
data = data.drop('name', 1)
data = data.drop('collection_id', 1)
```

```python
states = []
for i in data["state"]:
    states.append(i)
states = list(set(states))
states.sort()
for j, k in enumerate(data["state"]):
    data.at[j, "state"] = states.index(data.at[j, "state"])
```

```python
states = []
for i in data["city"]:
    states.append(i)
states = list(set(states))
states.sort()
for j, k in enumerate(data["city"]):
    data.at[j, "city"] = states.index(data.at[j, "city"])
```

```python
plan = {"Bronze": 1, "Silver": 2, "Gold": 3, "Platinum": 4, np.nan: 0}
for j, k in enumerate(data["PURCHASED"]):
    data.at[j, "PURCHASED"] = plan[data.at[j, "PURCHASED"]]
```

```python
for j, k in enumerate(data["DOB"]):
    data.at[j, "DOB"] = datetime.datetime.now().year - int(data.at[j, "DOB"][:4])
data.rename(columns={'DOB' : 'age'}, inplace=True)
data.columns
```

```python
sex = {"M": 0, "F": 1}
for j, k in enumerate(data["sex"]):
    try:
        data.at[j, "sex"] = sex[data.at[j, "sex"]]
    except KeyError:
        data = data.drop(j)
```

```python
tobacco = {"No": 0, "Yes": 1}
for j, k in enumerate(data["TOBACCO"]):
    try:
        data.at[j, "TOBACCO"] = tobacco[data.at[j, "TOBACCO"]]
    except KeyError:
        try:
            data = data.drop(j)
        except ValueError:
            pass
```

```python
status = {"S": 0, "M": 1}
for j, k in enumerate(data["MARITAL_STATUS"]):
    try:
        print(data.at[j, "MARITAL_STATUS"])
        data.at[j, "MARITAL_STATUS"] = status[data.at[j, "MARITAL_STATUS"]]
    except KeyError:
        try:
            data = data.drop(j)
        except ValueError:
            pass
```

```python
for i in data.columns:
    for k in range(0, len(data[i])):
        try:
            data.at[k, i] = int(data.at[k, i])
        except KeyError:
            pass
```

```python
the_data = data.values.tolist()
```

```python
for i in the_data:
    for k in i:
        if type(k) == str:
            print(k) 
```

```python
model = knn.NearestNeighbors(5)
model.fit(the_data)
```
