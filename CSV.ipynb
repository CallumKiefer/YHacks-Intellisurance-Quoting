```python
import re
import csv
import json
import numpy as np
import pandas
import collections
```

```python
data = pandas.read_csv("data2.csv", sep=",", header=0)
```

```python
data 
```

```python
data["PRE_CONDITIONS"][0]
```

```python
for j, k in enumerate(data["PRE_CONDITIONS"]):
    PRE_CONDITIONS = []
    try:
        preconditions = re.findall(r"['{'][A-Za-z\"_:\s,0-9.\\]+['}']", k)
        for i in preconditions:
            PRE_CONDITIONS.append(json.loads(i))
        data.at[j, "PRE_CONDITIONS"] = PRE_CONDITIONS
    except TypeError:
        pass
```

```python
ICD_codes = []
for i in data["PRE_CONDITIONS"]:
    for j in i:
        ICD_codes.append(j["ICD_CODE"])
```

```python
ctr = collections.Counter(ICD_codes)
```

```python
ctr
```

```python
d = pandas.DataFrame(np.zeros((len(data), len(ctr))))
d.columns = ctr
```

```python
data = pandas.concat([data, d], axis=1)
```

```python
risk = {"Low": 1, "Medium": 2, "High": 3}
for j, k in enumerate(data["PRE_CONDITIONS"]):
    for i in k:
        data.at[j, i["ICD_CODE"]] = risk[i["Risk_factor"]]
```

```python
data = data.drop('PRE_CONDITIONS', 1)
```
